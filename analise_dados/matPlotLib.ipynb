```python
import pandas as pb
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pb.read_csv('./gas_price/gas_prices.csv')
gas
```

```python
plt.title('Price in US$')

A = plt.plot(gas['Year'], gas['Australia'], 'r.-', label = 'Australia')
B = plt.plot(gas['Year'], gas['Italy'], 'g.-', label = 'Italy')
C = plt.plot(gas['Year'], gas['USA'], 'b.-', label = 'USA')

plt.xticks(gas['Year'][::+3])
plt.xlabel('Years')
plt.ylabel('US$')

plt.legend()
plt.show()
```

```python
list_country = ['Australia', 'USA', 'Mexico']

for country in gas:
    if country in list_country:
        plt.title('Price in US$')
        plt.xlabel('Years')
        plt.ylabel('US$')
        plt.plot(gas['Year'], gas[country], label=country)

plt.xticks(gas['Year'][::+3])
plt.legend()
plt.savefig('Country.pdf')
plt.savefig('Country.png')
plt.show()
```
